```python
pattern='\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages=re.split(pattern,data)[1:]
```

```python
messages
```

```python
dates=re.findall(pattern,data)
```

```python
dates
```

```python
df=pd.DataFrame({"user_message":messages,"message_date":dates})
```

```python
df
```

```python
df['message_date']
```

```python
df['message_date']=pd.to_datetime(df['message_date'],format="%m/%d/%y, %H:%M - ")
```

```python
df['message_date']
```

```python
df.shape
```

```python
df.rename(columns={'message_date':'date'},inplace=True)
```

```python
df
```

```python
re.split('([\w\W]+?):\s',df['user_message'][2])[1:]
```

```python
# we are going to split the user_message column into two columns of user and message
users=[]
messages=[]
for message in df['user_message']:
    entry=re.split("([\w\W]+?):\s",message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append("group notification")
        messages.append(entry[0])
df['user']=users
df['message']=messages
```

```python
df
```

```python
#now drop the user_message column from the df
df.drop('user_message',axis=1,inplace=True)
```

```python
f=open("stop_hinglish.txt",'r')
stop_words=f.read()
print(stop_words)
```

```python
#calculating the top 25 words
words=[]
for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)
words
```
